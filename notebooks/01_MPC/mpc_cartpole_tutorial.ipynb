# MPC Cart-Pole Tutorial

This notebook introduces a simple linear MPC formulation and compares it with LQR.

## 1. Linearized dynamics (toy model)

```
import numpy as np
import matplotlib.pyplot as plt

# Toy linear system
A = np.array([[0,1,0,0],[0,0,-0.1818,0],[0,0,0,1],[0,0,2.6727,0]])
B = np.array([[0],[1.818],[0],[-2.727]])
print('A matrix:\n', A)
```

## 2. LQR baseline (use `src.controllib.lqr`)

```
from src.controllib.lqr import lqr
Q = np.eye(4)
R = np.array([[0.01]])
K = lqr(A,B,Q,R)
print('LQR gain K =', K)
```
